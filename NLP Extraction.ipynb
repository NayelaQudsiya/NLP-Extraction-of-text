```python
import PyPDF2 as pdf
```

```python
file = open('E FRANCHISE SYSTEM REPORT 2.pdf', 'rb')
```

```python
## To know wheter the pdf is 
if pdf_reader.is_encrypted:
    print("The PDF is encrypted.")
else:
    print("The PDF is not encrypted.")
```

```python
len(pdf_reader.pages)
```

```python
for page_num in range(len(pdf_reader.pages)):
    page = pdf_reader.pages[page_num]
    print(f"Page {page_num + 1} text:")
    print(page.extract_text())
    print("\n")
```

```python
# Access the 3rd page (index 2)
page9 = pdf_reader.pages[8]

# Extract and print text from the 3rd page
print("Page 9 text:")
print(page9.extract_text())
```

### Append Write or Merge PDFs

```python
page9 = pdf_reader.pages[8]
page10 = pdf_reader.pages[9]
```

```python
pdf_writer = pdf.PdfWriter()
```

```python
# Access the first page
page1 = pdf_reader.pages[0]

pdf_writer.add_Page(page1)

# Save the new PDF
with open("output.pdf", "wb") as output_file:
    pdf_writer.write(output_file)

print("Page added to new PDF successfully!")
```

```python
pdf_writer.add_page(page9)
pdf_writer.add_page(page10)
```

```python
output = open('Pages.pdf','wb')
pdf_writer.write(output)
output.close()
```
